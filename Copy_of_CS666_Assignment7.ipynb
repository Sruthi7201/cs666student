**Task 4**: Let's apply a 3x3 filter mask! [30 Points]

```python
# Here we will modify the pixel data!
#
# The following code allows you to set the pixels of the current slice.
#
#   pixels.set(new_pixels);
#
# Write code to APPLY A FILTER MASK to the pixels.
#
# Hint: You can use the following code to get the dimensions of the current slice.
#   w = element.image.width;
#   h = element.image.height;
#
# You can decide which filter mask to apply (e.g., Sobel, Gaussian blur, edge detection..).
#
# After setting new pixel values, you can re-draw/re-render the current slice with this command:
#
#   cornerstone.renderGrayscaleImage(element, true);
#
```

```python

# #
# #
# # TODO: YOUR CODE GOES HERE
# #
# #
var element = cornerstone.getEnabledElements()[0];
var pixels = element.image.getPixelData();
var width = element.image.width;
var height = element.image.height;

// Create a new array to store the filtered pixel values
var filteredPixels = new Int16Array(pixels.length);

// Apply the averaging filter mask to each pixel
for (var y = 1; y < height - 1; y++) {
  for (var x = 1; x < width - 1; x++) {
    var pixelIndex = y * width + x;
    var sum = 0;

    // Calculate the average value of the 3x3 neighborhood
    for (var j = -1; j <= 1; j++) {
      for (var i = -1; i <= 1; i++) {
        var neighborIndex = (y + j) * width + (x + i);
        var neighborValue = pixels[neighborIndex];
        sum += neighborValue;
      }
    }

    var averageValue = Math.round(sum / 9); // Divide by the number of pixels in the neighborhood

    // Set the filtered pixel value
    filteredPixels[pixelIndex] = averageValue;
  }
}

// Set the filtered pixel values to the current slice
pixels.set(filteredPixels);

// Re-render the current slice
cornerstone.renderGrayscaleImage(element, true);
```

Bookmarklets and kernel mask filtering are two different ways of working with images on web pages.
Bookmarklets are small pieces of code that can change the web page content and do things like adding effects or changing shapes to images.
Kernel mask filtering is a method that changes each pixel of an image based on a formula that uses nearby pixels.
Bookmarklets only work on web pages and use JavaScript to get and change pixels through the web page structure.
Kernel mask filtering can work on any image and use any programming language or tool that supports image processing.
